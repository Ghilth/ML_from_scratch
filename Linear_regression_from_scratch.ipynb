```python
import numpy as np

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt

np.random.seed(42)
```

# Model from scratch

Y = X.a +b

```python
class Linear_regression:

  def __init__(self, X:np.ndarray, Y:np.ndarray, learning_rate: float, epoch:int):

    #get features and target values
    self.X = X
    self.Y = Y

    #get features array dimensions
    self.m= X.shape[0]
    self.n= X.shape[1]

    #learning_rate and epoch
    self.learning_rate = learning_rate
    self.epoch=epoch

    #a and b vectors initialisation
    self.a= np.random.randn(self.n,1)
    self.b= np.random.randn(self.m,1)




  def fit(self):

    for i in range(self.epoch):

      #gradient
      grad_a, grad_b=0, 0

      for j in range(self.m):

        grad_a=  grad_a + self.Y[j] - (self.X[j].dot(self.a) + self.b[j]).dot(self.X[j].reshape(1,self.n))
        grad_b = grad_b + self.Y[j] - (self.X[j].dot(self.a) + self.b[j])


      grad_a= - (grad_a/self.m).reshape(self.n,1)
      grad_b=  - grad_b/self.m


      #update
      self.a= self.a- self.learning_rate*grad_a
      self.b= self.b- self.learning_rate*grad_b


      #prediction
      loss=0.0
      for line in range(self.m):
        loss= loss + (self.Y[line] - (self.X[line].dot(self.a) + self.b[line]))**2
      loss =loss/(2*self.m)



      print(f"MSE : {loss} -------------------- epoch : {i+1}")



  def prediction(self,X):
    return X.dot(self.a) +self.b

```

```python
x=np.random.randn(1000,5)
y=np.random.randn(1000,1)


model= Linear_regression(x,y,0.5, 1000)
print(model.fit())
```

```python
y_p =model.prediction(x)


# Évaluer les performances du modèle
mse = mean_squared_error(y, y_p)

print("\nÉvaluation du modèle :")
print(f"Erreur quadratique moyenne (MSE) : {mse:.2f}")
```
