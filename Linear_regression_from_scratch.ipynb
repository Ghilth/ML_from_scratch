# Model

Y = X.a +b

```python
def model(X, y, lr, epoch):

  """"
  X : numpy array of features; X.shape =(n, p+1)
  y:  numpy array of target values; Y.shape=(n, 1)
  lr: learning rate for gradient descent

  """
  n,p= X.shape

  #add a colimn of 1 to X vector
  X= np.concatenate((np.ones([n,1]),X), axis=1)

  #Create a vector a
  a = np.random.rand(p+1, 1)

  #create a vector b
  b=np.random.rand(n,1)


  #descient gradient

  for i in range(epoch):

    grad_a=np.zeros([p+1,1])
    grad_b=np.zeros([p+1,1])


    for j in range(p+1):

      y_hat=X[j].dot(a) + b

      #loss
      loss= (y[j] - y_hat)**2
      loss= np.sum(loss)/2*(p+1)


      #gradient
      grad_a[j]= np.sum(loss*X[j])/p+1
      print(grad_a.shape)




      for j in range(p):


        y_hat=X[j].dot(a) + b

        #loss
        loss= (y[j] - y_hat)**2
        loss= np.sum(loss)/2*(n)


        #gradient
        grad_b[j]= np.sum(loss)/n
        print(grad_b.shape)


    a=a-lr*grad_a
    b=b-lr*grad_b

    print(f"a={a}, b={b}")

    return f"X.a+b with a={a} , b={b}, epoch={epoch} ans learning_rate={lr}"

```

```python
X=np.random.rand(50,7)
print(X)
```

```python
y=np.random.rand(50,1)
print(y)
```

```python
print(model(X,y,0.01,10))
```
